```python
fig, ax = plt.subplots(figsize=(12,8))

plt.scatter(df['Subscribers'], df['Video Views'], color='red', alpha=0.2)
plt.title("Relationship between Subscribers and Video Views", fontsize=20)
ax.set_facecolor('whitesmoke')
plt.xlabel('Number of Subscribers', fontsize=20)
plt.ylabel('Video Views', fontsize=20)
plt.tick_params(axis='both', labelsize=15)

;
```

```python
sns.set(style='whitegrid')

sns.scatterplot(x=df['Subscribers'], y=df['Video Views'], hue=df['Video Views'], cmap='coolwarm')
```

```python
MASK_FOR_2012 = (df_copy['Started'] == 2012)
```

```python
Top_25_2012 = df_copy[MASK_FOR_2012].sort_values("Subscribers")[50:]
Top_25_2012
```

```python
#Create plot
fig, ax = plt.subplots(figsize=(12,8))

ax.bar(Top_25_2012['Youtuber'], Top_25_2012['Subscribers'], color="red", alpha=0.3)
ax.set_xlabel("Youtuber Name", fontsize=16)
ax.set_ylabel("Subscriber Count (in 100 millions)", fontsize=16)
ax.set_title("Top 20 Youtubers from 2012 by Subscriber Count", fontsize=20)
ax.grid(axis='x')
plt.xticks(rotation=90)
;
```
